```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Retail Sales Analysis_utf.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
gender_count = sns.countplot(x = "gender" , data = df)
```

```python
df.rename(columns = {'quantiy':'Quantity'}, inplace = True)

df['total_cogs'] = df['cogs'] * df['price_per_unit']
gender_data = df.groupby('gender', as_index=False)[['Quantity', 'total_cogs', 'total_sale']].sum()
sns.barplot(x = 'gender',y= 'Quantity' ,data = gender_data)

```

```python
sns.barplot(x = 'gender',y= 'total_cogs' ,data = gender_data)
```

```python
sns.barplot(x = 'gender',y= 'total_sale' ,data = gender_data)
```

From the above three plots. We can clearly see that females are buying more quantity and therefore their purchasing power is relatively more than men

```python
bins = [0, 18, 25, 35, 45, 55, 65, 100]
label = ['<18', '18–25', '26–35', '36–45', '46–55', '56–65', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=label)
sns.countplot(data=df, x='age_group', hue='gender')
plt.xlabel('Age Group')
plt.ylabel('Customer Count')
plt.title('Customer Count by Age Group')
plt.show()
```

```python
sns.countplot(data=df, x='age_group', hue='category')
plt.xlabel('Age Group')
plt.ylabel('Customer Count')
plt.title('Customer Count by Age Group')
plt.show()
```

```python
category_data = df.groupby('category', as_index=False)['total_sale'].sum()
sns.barplot(x='category',y = 'total_sale' , data = category_data)
plt.xlabel('category')
plt.ylabel('Customer Count')
plt.title('Customer Count by category')
plt.show()
```

```python
sales_category = df .groupby(['category'], as_index=False)['total_sale'].sum().sort_values(by='total_sale', ascending=False).head(3)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(data = sales_category, x = 'category' , y= 'total_sale')
```

```python
sales_customer = df .groupby(['customer_id'], as_index=False)['total_sale'].sum().sort_values(by='total_sale', ascending=False).head(10)
sns.barplot(data = sales_customer, x = 'customer_id' , y= 'total_sale')
```
